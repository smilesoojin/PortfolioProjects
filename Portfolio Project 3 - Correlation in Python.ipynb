```python
# Import librarires

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)    # Adjusts the configuration of the plots we create

pd.set_option('display.max_rows', None)    # Display all rows - not the best practice but useful at early stage of data evaluation
```

```python
# Read in the data

df = pd.read_csv('D:\Documents\# 마리아\Coding\Portfolio Project 1\movies.csv')
df.head()
```

```python
# Check for missing data

for col in df.columns:
    percent_missing = np.mean(df[col].isnull())
    print('{}: {}%'.format(col, percent_missing))

# Findings: Percents are relatively low. No need to worry about missing data for now.
```

```python
# Looking at data types for each column

df.dtypes
```

```python
# Adjusting data types

df = df.dropna()    # Gives error without this statement: cannot convert non-finite types

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
# 'Year' and 'Year Released' don't match for all rows
# Create new year column from release date

df['year_released'] = df['released'].str.extract(pat='([0-9]{4})').astype(int)
```

```python
# Check for duplicates in Company names

#df['company'].drop_duplicates().sort_values(ascending=False).head()

# Findings: No need to adjust company names as they all represent different firms
#           (e.g. Universal Pictures International is different from Universal Pictures)
```

```python
# Sort dataframe by gross value to brainstorm what factors may impact higher gross

df.sort_values(by=['gross'], inplace=False, ascending=False).head()
```

```python
# Scatter plot with gross vs. budget using matplotlib

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Gross Earnings vs. Budget')
plt.xlabel('Budget for Film')
plt.ylabel('Gross Earnings')

plt.show()
```

```python
# Plot gross vs. budget using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

```python
# Evaluate Correlation Coefficients (this only works on numerical values)

df.corr()   # methods: pearson (default), kendall, spearman

# Findings: High correlation between budget and gross (value closest to 1)
```

```python
# Visualize above findings using correlation matrix

correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# Above analysis is only applicable to numeric features
# Convert non-numeric features to numeric values

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes    # gives random numerical values

df_numerized.head()
```

```python
# Recreate correlation matrix using numerized data

correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# Correlation Matrix using numerized data

df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs.head()
```

```python
sorted_pairs = corr_pairs.sort_values(ascending=False)
sorted_pairs.head()
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
